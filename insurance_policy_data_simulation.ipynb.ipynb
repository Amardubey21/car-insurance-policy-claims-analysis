```python
import pandas as pd
import numpy as np
```

```python
customers = 1000000

dates = pd.date_range("2024-01-01","2024-12-31")

purchase_dates = np.random.choice(dates,customers)

tenure = np.random.choice(
    [1,2,3,4],
    size=customers,
    p=[0.2,0.3,0.4,0.1]
)

policy = pd.DataFrame({
"Customer_ID": range(1,customers+1),
"Vehicle_ID": range(1000001,1000001+customers),
"Vehicle_Value":100000,
"Policy_Tenure":tenure,
"Policy_Purchase_Date":purchase_dates
})

policy["Premium"]=policy["Policy_Tenure"]*100
policy["Policy_Start_Date"]=policy["Policy_Purchase_Date"]+pd.Timedelta(days=365)
policy["Policy_End_Date"]=policy["Policy_Start_Date"]+pd.to_timedelta(policy["Policy_Tenure"]*365,unit="D")

policy.to_csv("policy_sales_data.csv",index=False)

policy.head()
```

```python
policy = pd.read_csv("policy_sales_data.csv")

policy["Policy_Purchase_Date"]=pd.to_datetime(policy["Policy_Purchase_Date"])
policy["Policy_Start_Date"]=pd.to_datetime(policy["Policy_Start_Date"])

claim_days=[7,14,21,28]

eligible = policy[policy["Policy_Purchase_Date"].dt.day.isin(claim_days)]

claims_2025 = eligible.sample(frac=0.30)

claims_2025["Claim_Date"]=claims_2025["Policy_Start_Date"]
claims_2025["Claim_Type"]=1
claims_2025["Claim_Amount"]=10000

four_year = policy[policy["Policy_Tenure"]==4]

claims_2026 = four_year.sample(frac=0.10)

dates = pd.date_range("2026-01-01","2026-02-28")

claims_2026["Claim_Date"]=np.random.choice(dates,len(claims_2026))
claims_2026["Claim_Type"]=2
claims_2026["Claim_Amount"]=10000

claims=pd.concat([claims_2025,claims_2026])

claims["Claim_ID"]=range(1,len(claims)+1)

claims = claims[[
"Claim_ID",
"Customer_ID",
"Vehicle_ID",
"Claim_Amount",
"Claim_Date",
"Claim_Type"
]]

claims.to_csv("claims_data.csv",index=False)

claims.head()
```
